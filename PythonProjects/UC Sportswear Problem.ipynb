# UC Sportswear Problem

```python
import random as rd
import statistics as st
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.stats import *
init_printing()
```

```python
# Imput parameters
Shirt_cost = 50.00 # $/shirt
Shirt_sell_price = 75.00 # $/shirt
leftover_shirts = 5 # shirts
leftover_price = 15.00 # $/shirt
Average_sales = 50 # shirts
Shirt_order = 55 # shirts
```

```python
# Generate simulation sample
trials = 10000
sample = list()
for i in range(trials):
    demand = np.random.poisson(Average_sales)
    if demand < Shirt_order:
        qty_sold = demand
    else:
        qty_sold = Average_sales
    qty_left = Average_sales- qty_sold
    Sales_revenue = Shirt_sell_price * qty_sold
    leftover_revenue = leftover_price * qty_left
    total_cost = Shirt_cost * Shirt_order
    profit = Sales_revenue + leftover_revenue - total_cost
    sample.append(profit)
```

```python
plt.hist(sample, edgecolor = 'k')
```

```python
# Analyize Sample Statistics

print('Max profit =', max(sample))
print('Mode profit=', st.mode(sample))
print('Median profit =', st.median(sample))
print('Mean profit =', st.mean(sample))
print('Min profit =', min(sample))
```

```python
# Probility for Net Loss
proportion = sum(1 for x in sample if x < 0)/len(sample)
print('Probility for net loss is', proportion)
```
